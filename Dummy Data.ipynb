```python
import pandas as pd
import urllib.request as curl
import json
import pprint

event_url = "http://www.ausstage.edu.au/exchange/events?type=organisation&id=590&output=json&limit=all"
place_url = "https://www.ausstage.edu.au/opencms/markers?type=organisation&id=590"

def request_and_parse_json(url_string):
    data_event = curl.urlopen(url_string).read()
    JSON_object = json.loads(data_event.decode("utf-8","ignore"))
    return JSON_object

def format_event_place(string):
    string = string.split(",")
    if len(string) > 0:
        return string[0]
    return ""
```

```python
event_raw = request_and_parse_json(event_url)
events = pd.DataFrame(event_raw)
events = events[1:]
events = events.drop('_generator', axis=1)
print(events)
```

```python
place_raw = request_and_parse_json(place_url)
place = pd.DataFrame(place_raw[0]['venues'])
print(place)
```

```python
events['venue'] = events['venue'].apply(format_event_place)
print(events)
```

```python
print(place[place['name'] == "Theatre Royal (1976- )"])
```

```python
merged = pd.merge(events, place, left_on = 'venue', right_on = 'name')
```
